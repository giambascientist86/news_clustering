## Now to Vectorization methods comparison:


```python
import sys
sys.path.append(r"C:\Users\batti\topic-modeling")

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from src.utils.vectorization import Vectorizer

# Load preprocessed dataset
df = pd.read_csv("../data/cleaned_news.csv")

# Initialize Vectorizer
vectorizer = Vectorizer()

# Choose vectorization methods
methods = ["tfidf", "word2vec", "bert", "doc2vec"]
embeddings = {}

# Compute embeddings
for method in methods:
    embeddings[method] = vectorizer.vectorize(df, method=method)

# Dimensionality Reduction using PCA
def reduce_dimensionality(embedding_matrix, method="PCA"):
    if method == "PCA":
        reducer = PCA(n_components=2)
    elif method == "TSNE":
        reducer = TSNE(n_components=2, perplexity=30, random_state=42)
    return reducer.fit_transform(embedding_matrix)

# Visualization
fig, axes = plt.subplots(2, 2, figsize=(12, 12))
axes = axes.flatten()

for i, method in enumerate(methods):
    reduced_data = reduce_dimensionality(embeddings[method], method="PCA")
    
    sns.scatterplot(x=reduced_data[:, 0], y=reduced_data[:, 1], ax=axes[i])
    axes[i].set_title(f"{method.upper()} Embeddings (PCA Reduced)")

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import scipy.sparse
import matplotlib.pyplot as plt
import seaborn as sns

# Path al file TF-IDF salvato
tfidf_path = r"C:\Users\batti\topic-modeling\data\tfidf.npz"
feature_names_path = tfidf_path.replace(".npz", "_features.npy")

# Caricamento della matrice TF-IDF
tfidf_matrix = scipy.sparse.load_npz(tfidf_path)
feature_names = np.load(feature_names_path, allow_pickle=True)

# Esplorazione base
print(f"Shape della matrice TF-IDF: {tfidf_matrix.shape}")
print(f"Tipo della matrice: {type(tfidf_matrix)}")

# Convertiamo una parte della matrice in densa per visualizzarla
sample_matrix = tfidf_matrix[:10].toarray()

# Creazione di una heatmap per visualizzare la densità
plt.figure(figsize=(12, 6))
sns.heatmap(sample_matrix, cmap="Blues", cbar=True, xticklabels=False, yticklabels=False)
plt.title("Densità della Matrice TF-IDF (primi 10 documenti)")
plt.show()

# Verifica della sparseness
total_elements = tfidf_matrix.shape[0] * tfidf_matrix.shape[1]
nonzero_elements = tfidf_matrix.count_nonzero()
density = nonzero_elements / total_elements
print(f"Densità della matrice TF-IDF: {density:.6f}")

# Mostra le prime parole più frequenti
word_sums = np.asarray(tfidf_matrix.sum(axis=0)).flatten()
sorted_indices = np.argsort(word_sums)[::-1]
top_words = [feature_names[i] for i in sorted_indices[:10]]
top_values = word_sums[sorted_indices[:10]]

plt.figure(figsize=(10, 5))
plt.barh(top_words[::-1], top_values[::-1], color='skyblue')
plt.xlabel("Somma TF-IDF")
plt.title("Top 10 Parole più Rilevanti nella Matrice TF-IDF")
plt.show()
```
